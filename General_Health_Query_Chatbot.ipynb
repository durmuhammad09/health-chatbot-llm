```python
# ==========================================
# General Health Query Chatbot
# Working Google Colab Version
# ==========================================


# ------------------------------------------
# STEP 1: Install Libraries
# ------------------------------------------

!pip install -q transformers torch sentencepiece


# ------------------------------------------
# STEP 2: Import Libraries
# ------------------------------------------

from transformers import pipeline


# ------------------------------------------
# STEP 3: Load AI Model
# ------------------------------------------

print("Loading AI model...")

chatbot = pipeline(
    "question-answering"
)

print("Model loaded successfully!")


# ------------------------------------------
# STEP 4: Unsafe Keywords
# ------------------------------------------

unsafe_keywords = [
    "suicide",
    "kill",
    "overdose",
    "self-harm",
    "poison"
]


# ------------------------------------------
# STEP 5: Safety Filter Function
# ------------------------------------------

def is_safe(question):

    question = question.lower()

    for word in unsafe_keywords:

        if word in question:
            return False

    return True


# ------------------------------------------
# STEP 6: Health Chatbot Function
# ------------------------------------------

def health_chatbot(question):

    # Safety check
    if not is_safe(question):

        return "Sorry, I cannot provide harmful medical advice."

    # Health knowledge context
    context = """
    A sore throat can happen because of viral infections,
    colds, flu, allergies, smoking, or bacteria.

    Paracetamol is usually safe for children when used
    in the correct dosage recommended by a doctor.

    Fever symptoms may include high temperature,
    sweating, weakness, headache, and chills.

    Drinking water helps keep the body hydrated.

    Healthy eating and exercise improve health.
    """

    # Generate answer
    response = chatbot(
        question=question,
        context=context
    )

    return response["answer"]


# ------------------------------------------
# STEP 7: Start Chatbot
# ------------------------------------------

print("\n===== General Health Chatbot =====")
print("Type your question")
print("Type 'quit' to stop chatbot\n")

while True:

    # Ask user question
    user_question = input("Ask your health question: ")

    # Exit condition
    if user_question.lower() == "quit":

        print("\nChatbot ended.")
        break

    # Get chatbot response
    answer = health_chatbot(user_question)

    # Print response
    print("\nChatbot Response:")
    print(answer)
    print("\n")
```
